*1. Pivot across Categories *  
The ebola data currently has a single column called category which contains the following types of Ebolas cases:
- Cases
- Deaths
- Suspected Cases
- Confirmed Cases
- Probable Cases

We want to pivot the row values into distinct columns. That is we want to create 5 new columns for each of category types

```python
#1. Pivot across these category types
ebola_pivot = ebola.pivot_table('Value', ['Country','Date'], 'Category', aggfunc = np.mean)
ebola = ebola_pivot.reset_index()   #Reset DF index

#2. fill missing values is nan
ebola.fillna('nan')
```

```python
#Out CSV

ebola = pd.read_csv("data/ebola_clean.csv")
#Build Json
json = {}
cols = ebola.columns
for r in range(0,len(ebola)):
    json[r] = {}
    for col in cols:
        if (ebola.irow(r)[col] != ebola.irow(r)[col]):
            json[r][col] = "nan"
        else:
            json[r][col] = ebola.irow(r)[col]
print json
```
